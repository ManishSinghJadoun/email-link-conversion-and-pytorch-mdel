```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, roc_auc_score
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader

# Load your dataset
df = pd.read_csv("updated_email_table.csv")

# Convert target to binary
df['link_clicked'] = df['link_clicked'].map({'Yes': 1, 'No': 0})
df['email_opened'] = df['email_opened'].map({'Yes': 1, 'No': 0})

# Encode categorical variables
for col in ['email_version', 'email_text', 'weekday', 'user_country']:
    df[col] = LabelEncoder().fit_transform(df[col])

# Features and target
X = df[['email_version', 'email_text', 'hour', 'weekday', 'user_country', 'user_past_purchases']]
y = df['link_clicked']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = X_scaled.reshape((X_scaled.shape[0], 1, X_scaled.shape[1]))

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Convert to tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32).unsqueeze(1)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.float32).unsqueeze(1)

# Dataloaders
train_data = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_data, batch_size=64, shuffle=True)

# Define the RNN model
class EmailRNN(nn.Module):
    def __init__(self):
        super(EmailRNN, self).__init__()
        self.lstm = nn.LSTM(input_size=6, hidden_size=32, batch_first=True)
        self.fc1 = nn.Linear(32, 16)
        self.dropout = nn.Dropout(0.3)
        self.fc2 = nn.Linear(16, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        _, (h_n, _) = self.lstm(x)
        x = torch.relu(self.fc1(h_n[-1]))
        x = self.dropout(x)
        x = self.fc2(x)
        return self.sigmoid(x)

# Instantiate model
model = EmailRNN()
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Train the model
for epoch in range(20):
    model.train()
    for batch_X, batch_y in train_loader:
        optimizer.zero_grad()
        outputs = model(batch_X)
        loss = criterion(outputs, batch_y)
        loss.backward()
        optimizer.step()
    print(f"Epoch {epoch+1}/20 - Loss: {loss.item():.4f}")

# Evaluate
model.eval()
with torch.no_grad():
    y_proba = model(X_test_tensor).squeeze().numpy()
    y_pred = (y_proba > 0.5).astype(int)
    auc = roc_auc_score(y_test, y_proba)
    acc = accuracy_score(y_test, y_pred)
    print(f"AUC: {auc:.4f}, Accuracy: {acc:.4f}")

    # Simulate top 20% targeting
    top_20_idx = np.argsort(y_proba)[-int(0.2 * len(y_proba)):]
    print("Model-based expected CTR (top 20%):", y_test.iloc[top_20_idx].mean())

# Save model and scaler
import joblib
torch.save(model.state_dict(), "email_rnn_model.pt")
joblib.dump(scaler, "scaler.pkl")

```

```python
# inference.py
import torch
import numpy as np
import pandas as pd
import joblib
from email_model import EmailRNN  # define model class as in training

def predict_click_prob(new_data_df):
    scaler = joblib.load("scaler.pkl")
    model = EmailRNN()
    model.load_state_dict(torch.load("email_rnn_model.pt"))
    model.eval()

    # Preprocess
    new_data = scaler.transform(new_data_df)
    new_data = new_data.reshape((new_data.shape[0], 1, new_data.shape[1]))
    new_tensor = torch.tensor(new_data, dtype=torch.float32)

    # Predict
    with torch.no_grad():
        predictions = model(new_tensor).squeeze().numpy()

    return predictions
```
